```python
df = df.reset_index()
```

```python
avg_height = [0]*len(df)
avg_weight = [0]*len(df)
for i in range(len(df)):
  avg_height[i] = (df['height_high_inches'][i]+df['height_low_inches'][i])/2
  avg_weight[i] = (df['weight_high_lbs'][i]+df['weight_low_lbs'][i])/2
```

```python
df['avg_height'] = avg_height
df['avg_weight'] = avg_weight
```

```python
df.reset_index()
h_var = [0]*len(df)
w_var = [0]*len(df)
reps_var = [0]*len(df)
for i in range(len(df)):
  h_var[i] = (df['height_high_inches'][i]-df['height_low_inches'][i])/(df['height_high_inches'][i]+df['height_low_inches'][i])
  w_var[i] = (df['weight_high_lbs'][i]-df['weight_low_lbs'][i])/(df['weight_high_lbs'][i]+df['weight_low_lbs'][i])
  reps_var[i] = (df['reps_upper'][i]-df['reps_lower'][i])/(df['reps_upper'][i]+df['reps_lower'][i])
```

```python
df['h_var'] = h_var
df['w_var'] = w_var
df['reps_var'] = reps_var
```
